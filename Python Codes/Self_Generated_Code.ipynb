```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer

# Set a random seed for reproducibility
np.random.seed(903949505)

positive_class_percentage = 20
class_weights = [positive_class_percentage / 100, (100 - positive_class_percentage) / 100]

# Create a synthetic dataset with binary classification and 50 variables
X, y = make_classification(
    n_classes=2,
    n_samples=10000,      # Number of samples
    n_features=100,       # Number of features (variables)
    n_informative=90,    # Number of informative features
    n_redundant=10,      # Number of redundant features
    n_clusters_per_class=3,  # Number of clusters per class
    flip_y=0.1,          # Fraction of samples whose class is assigned randomly
    random_state=903949505,      # Random seed
    weights=class_weights,
    shift=None,
    scale=None
)

# Add noise to the dataset
noise = np.random.choice([0, 10, -10], size=X.shape)
replace_percentage = 95
mask = np.random.choice([True, False], noise.shape, p=[replace_percentage / 100, (100 - replace_percentage) / 100])
noise[mask] = 0
X += noise

# Specify the indices of columns to apply polynomial features
columns_to_transform = [10,20,30,40,50,60,70,80,90]  # Replace with your desired column indices

# Create a ColumnTransformer to apply PolynomialFeatures only to specified columns
preprocessor = ColumnTransformer(
    transformers=[
        ('poly', PolynomialFeatures(degree=2, include_bias=False), columns_to_transform)
    ],
    remainder='passthrough'  # Include the remaining columns as they are
)

# Apply the ColumnTransformer to the dataset
X_transformed = preprocessor.fit_transform(X)

# Combine features and labels into a DataFrame
columns = [f"feature_{i}" for i in range(1, X_transformed.shape[1] + 1)]
df = pd.DataFrame(np.hstack([X_transformed, y.reshape(-1, 1)]), columns=columns + ["target"])
```
